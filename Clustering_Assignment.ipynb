```python
import numpy as np
from sklearn import cluster, datasets
import scipy.misc
from scipy import misc
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
racoon=scipy.misc.face(gray=True)                 
print("shape of face is:",racoon.shape)
print("data type of face is:",racoon.dtype)
```

```python
n_clusters=5
np.random.seed(0)
X=racoon.reshape((-1,1))
X.shape
```

```python
# Implementing K-means clustering to form K clusters
kmeans=KMeans(n_clusters=n_clusters,n_init=10)
kmeans.fit(X)
```

```python
values= kmeans.cluster_centers_.squeeze()
labels =kmeans.labels_
print(values)
print(labels)
```

```python
CompressedImage=np.choose(labels,values)
CompressedImage.shape=racoon.shape
vmin=racoon.min()
vmax=racoon.max()
```

```python
# Raw Image
plt.figure(1)
plt.imshow(racoon,cmap=plt.cm.gray,vmin=vmin,vmax=256)
plt.title('Raw Image')
# Compressed Image
plt.figure(2)
plt.imshow(CompressedImage,cmap=plt.cm.gray,vmin=vmin,vmax=256)
plt.title('Compressed Image')
```
